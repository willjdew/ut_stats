Generate a histogram for the length of time these adopted animals were in the shelter.

```python
plt.hist(animal_adopted["Days.Shelter"], color = 'blue')
plt.show()
```

```python
from __future__ import division
import math
import sys
 
def fivenum(array):
    n = len(array)
    if n == 0:
        print("you entered an empty array.")
        sys.exit()
    x = sorted(array)
 
    n4 = math.floor((n+3.0)/2.0)/2.0
    d = [1, n4, (n+1)/2, n+1-n4, n]
    sum_array = []
 
    for e in range(5):
        floor = int(math.floor(d[e] - 1))
        ceil = int(math.ceil(d[e] - 1))
        sum_array.append(0.5 * (x[floor] + x[ceil]))
 
    return sum_array
```

Select the appropriate measures of center and spread to describe the distribution.

```python
print fivenum(animal_adopted["Days.Shelter"])
print animal_adopted["Days.Shelter"].mean()
print animal_adopted["Days.Shelter"].std()
```

```python
z_score = (211 - animal_adopted["Days.Shelter"].mean()) / animal_adopted["Days.Shelter"].std()
z_score
```

### Conclution
